### Use Categories and Sales.csv and find the amount people are spending on different categories of products  which finds which product is being costly and which is low.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('categories.csv', delimiter=';')
df2=pd.read_csv('sales.csv', delimiter=';')
df3=pd.read_csv('products.csv', delimiter=';')
```

```python
df4=pd.DataFrame()
df4['ProductID']=df2['ProductID']
df4['Quantity']=df2['Quantity']
df4['Discount']=df2['Discount']
```

```python
df4
```

```python
df4['Discount'].isna().sum()
```

```python
df4['Discount'].fillna(0,inplace=True)
```

```python
df4.head(5)
```

```python
out=(df4.merge(df3,left_on='ProductID',right_on='ProductID').reindex(columns=['ProductID','Quantity','Price','Discount','CategoryID']))
```

```python
out
```

```python
final=(out.merge(df,left_on='CategoryID',right_on='CategoryID').reindex(columns=['ProductID','Quantity','Price','Discount','CategoryID','CategoryName']))
```

```python
final
```

```python
final['Price']=final['Price'].astype(int)
```

```python
# fill the missing values with 100 
final['Price']=final['Quantity'].mul() 
```

```python
final['final TD']=final['Quantity']*final['Discount']
```
