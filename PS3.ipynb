### Which age group of employees are selling the most products and create a pie chart to visualize it.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('employes.csv',delimiter=';')
df.head()
```

```python
df['BirthDate']=df['BirthDate'].str.split(' ',n=1,expand=True)
```

```python
df
```

```python
from datetime import datetime
from datetime import date
def Age(born):
    born=datetime.strptime(born,'%Y-%m-%d').date()
    today=date.today()
    return today.year-born.year-((today.month,today.day)<(born.month,born.day))
df['Age']=df['BirthDate'].apply(Age)
print(df)
```

```python
df2=pd.DataFrame()
df2['EmployeeID']=df['EmployeeID']
df2['Gender']=df['Gender']
df2['Age']=df['Age']
df2['Name']=df['FirstName']+" "+df['LastName']
```

```python
df2
```

```python
df3=pd.read_csv('sales.csv',delimiter=';')
df3.head(2)
df3.shape
```

```python
df4=df3['SalesPersonID'].value_counts().rename_axis('EmployeeID').to_frame(name='count')
```

```python
df4
```

```python
final=(df4.merge(df2,left_on='EmployeeID',right_on='EmployeeID').reindex(columns=['EmployeeID','count','Age','Gender','Name']))
```

```python
final
```

```python
sizes=final['count']
labels=final['Name']
plt.bar(labels,sizes,alpha=0.5,color='#123432')
plt.xticks(rotation='vertical')
plt.ylim(292000,295000)
```
